```python
from GoodReadsScraper import GoodReadsScraper

gr = GoodReadsScraper()
gr.createDriver()
books = gr.searchBook("All the President's Men")

for book in books:
    print(book.title)
    print(book.link)
    print(book.rating)
    print(book.rating_count)
    # print for author in authors, print author.name and author.link
    for author in book.authors:
        print(author.name, "–", author.link)
    print("")

gr.closeDriver()
```
